```python
#simple linear regression model
#importing necessary libraries
import pandas as pd #data manupulation
import numpy as np #linear algera and arrays
import matplotlib.pyplot as plt #visualization
```

```python
# load the datasets
df = pd.read_csv('appliance_energy.csv')
df.head() # to display top 5 rows - tail()-bottom 5 rows
```

```python
#tp know rows or columns
df.shape
```

```python
#information
df.info()
```

```python
#stastical info-datasets
df.describe()
```

```python
# to check the missing values
df.isnull().sum() #if we have missing values - drop/impute
```

```python
#feature and target selection
X = df[['Temperature (°C)']]
Y = df['Energy Consumption (kWh)']

```

```python
#split the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
#let's create a regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#train the model
model.fit(X_train,Y_train)
```

```python
#prediction
lr_y_pred = model.predict(X_test)
lr_y_pred
```

```python
#merrics
from sklearn.metrics import r2_score,mean_squared_error
mse = mean_squared_error(Y_test,lr_y_pred)
r2 = r2_score(Y_test,lr_y_pred)
print(f"the mean squared error is {mse}")
print(f"the r2 score is {r2}")
```

```python
#plot
import matplotlib.pyplot as plt
plt.scatter(X_test,Y_test,color='blue',label='Test Data')
plt.plot(X_test,lr_y_pred,color='red',label='Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Energy Consumption (kWh)')
plt.title('Energy Consumption using linear regression')
plt.show()
```

```python
#to save the model
import joblib
joblib.dump(model,'energy_consumption_model.pkl')
```

```python
#polynomial regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
#load the dataset
df = pd.read_csv('sustainability_data.csv')
df.tail()
```

```python
#to see column name
df.columns
```

```python
#feature selection
x = df[['Year', 'Energy_Consumption', 'Renewable_Percentage', 'GDP']]
y = df['CO2_Emissions']
```

```python
#split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
#define the degree of polynomial
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
#initialize the model
poly_model = LinearRegression()
#train the model
poly_model.fit(x_train_poly,y_train)
```

```python
#prediction
poly_pred =poly_model.predict(x_test_poly)
poly_pred
```

```python
#evaluate
mse = mean_squared_error(y_test,poly_pred)
r2 = r2_score(y_test,poly_pred)
print(f"the mean squared error is {mse}")
print(f"the r2 score is {r2}")
```

```python
import joblib
joblib.dump(poly_model,'sustainability_model.pkl')
```

```python
#classification
#importing necessory libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
#load the datasets
df = pd.read_csv('green_tech_data.csv')
df.head(10)#top 10 rows
```

```python
#find missing values
df.isnull().sum()
```

```python
#no of rows and columns
df.shape
```

```python
#col names
df.columns
```

```python
#features selection
x = df[['carbon_emissions', 'energy_output', 'renewability_index','cost_efficiency']]
y = df['sustainability']
```

```python
#split the datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#initialize and train the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
#train
model.fit(x_train,y_train)
```

```python
#prediction
y_pred = model.predict(x_test)
y_pred
```

```python
#metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy = accuracy_score(y_test,y_pred)
print(f"the accuracy is : {accuracy}")
```

```python
#CM
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True,cmap='coolwarm',xticklabels=['not sustainable','sustainable'],yticklabels=['not sustainable','sustainable'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
#classification report
cr = classification_report(y_test,y_pred,target_names=['not sustainable','sustainable'])
print(cr)
```

```python
# feature importance
co = pd.DataFrame(model.coef_.T,index=x.columns,columns=['coefficient'])
co
```

```python
#save model
import joblib
joblib.dump(model,'green_tech_model.pkl')
```

```python
x = df[['soil_health', 'crop_yield', 'water_usage', 'carbon_footprint','fertilizer_use']]
y = df['sustainability']
print(x.shape)
print(y.shape)
```

```python
#split the datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#initialize and train the model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100,random_state=42)
#train
model.fit(x_train,y_train)
```

```python
#prediction
y_pred = model.predict(x_test)
print('The prediction are',y_pred)
```

```python
#metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy = accuracy_score(y_test,y_pred)
print(f"the accuracy is : {accuracy}")
```

```python
#CM
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
cr = classification_report(y_test,y_pred,target_names=['not sustainable','sustainable'])
print(cr)
```
